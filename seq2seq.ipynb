```python
import pandas as pd
import numpy as np
import nltk
import os
from os import getcwd
from os import listdir
import string
```

```python
path='D:\\NLP_Practise\\Articles'
```

```python
headlines=[]
for filename in listdir(path):
    if 'Articles' in filename:
        article_df=pd.read_csv(path+f'\{filename}')
        headlines.extend(list(article_df['headline'].values))
headlines
```

```python
headlines=[head for head in headlines if head!='Unknown']
headlines
```

```python
head1=[]
for i in headlines:
    words=''.join([head for head in i if head not in string.punctuation]).lower()
    head1.append(words)
head1
```

```python
vocab=[]
for head in head1:
    words=head.split()
    vocab.extend(words)
vocabulary=set(vocab)
```
