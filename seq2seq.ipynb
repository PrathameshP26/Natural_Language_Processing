```python
import pandas as pd
import numpy as np
import nltk
import os
from os import getcwd
from os import listdir
import string
```

```python
path='D:\\NLP_Practise\\Articles'
```

```python
headlines=[]
for filename in listdir(path):
    if 'Articles' in filename:
        article_df=pd.read_csv(path+f'\{filename}')
        headlines.extend(list(article_df['headline'].values))
headlines
```

```python
headlines=[head for head in headlines if head!='Unknown']
headlines
```

```python
head1=[]
for i in headlines:
    words=''.join([head for head in i if head not in string.punctuation]).lower()
    words=words.encode("utf8").decode("ascii","ignore")
    head1.append(words)
head1
```

```python
vocab=[]
for head in head1:
    words=head.split()
    vocab.extend(words)
vocabulary=set(vocab)
```

```python
vocabulary
```

```python
len(vocabulary)
```

```python
from keras_preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(head1)
word2index=tokenizer.word_index
word2index
```

```python
dictionary={}
rev_dictionary={}
for key,value in word2index.items():
    dictionary[key]=value
    rev_dictionary[value]=key
```

```python
dictionary
```

```python
rev_dictionary
```

```python
input_sequences=tokenizer.texts_to_sequences(head1)
input_sequences
```

```python
input_data=[]
target=[]
for line in input_sequences:
    for i in range(1,len(line)-1):
        input_data.append(line[:i])
        target.append(line[i])

        
```

```python
input_data
```

```python
target
```

```python
MAXLEN=0
for i in input_data:
    if len(i)>MAXLEN:
        MAXLEN=len(i)
```

```python
MAXLEN
```

```python
input_data=pad_sequences(input_data,maxlen=MAXLEN,padding="post")
input_data
```

```python
input_data.shape
```

```python
target=to_categorical(target)
target
```

```python
import keras
```

```python
model=keras.models.Sequential()
model.add(keras.layers.Embedding(input_dim=len(vocabulary),output_dim=100,input_length=MAXLEN))
model.add(keras.layers.LSTM(units=100))
model.add(keras.layers.Dropout(rate=0.01))
model.add(keras.layers.Dense(units=target.shape[1],activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(input_data,target,batch_size=5,epochs=10)
```
